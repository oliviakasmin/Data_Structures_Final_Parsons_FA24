```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("scraped_data_uncleaned.csv")
```

```python
df
```

```python
cols_to_clean = ["National Priorities List Status", "Human Exposure Under Control", "Groundwater Migration Under Control"]
```

### Check if cols have the same pattern for all values in that col

```python
pattern = r"^[\n\t]+|[\n\t ]+$"

# test a few values
val1 = '\n\t\t\t\n\t\t\t\t\tFinal NPL\n\t\t\t\t\n\t\t'
# re.split(pattern, val1)

val2='\n\t\t\n\t\t\tNo\n\t\t\n\t'
re.split(pattern, val2)
```

```python
vals_dict = {}

for col in cols_to_clean:
    vals = df[col].unique().tolist()
    split_vals = map(lambda x: re.split(pattern, x)[1], vals)
    vals_dict[col] = list(split_vals)
```

```python
vals_dict
```

```python
for index, row in df.iterrows():
    for col in cols_to_clean:
        value = row[col]
        cleaned_value = re.split(pattern, value)[1]
        # print(cleaned_value)
        df.at[index, col] = cleaned_value
```

```python
for col in cols_to_clean:
    vals = df[col].unique().tolist()
    print(vals)
```
